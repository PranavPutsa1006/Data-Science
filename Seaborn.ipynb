Introduction: 
Seaborn Library:

Used for data Visualization and it is based on the Matplotlib.

It allows the creation of Statistical graphics

Functionalities:

Allows you to compare between the multiple variables

Supports the multiplot grids

different colour palettes are available in seaborn

Univariate and bivariate visualizations can be represented better.

Estimates and plot the Linear regression automatically

Why Seaborn?

Matplotlib settings are difficult to workout and user's responsibility is more. In seaborn we have customized themes and highlevel interfaces.

Seaborn functions work well on DataFrames.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Essential functions of Seaborn:

Visualizing statistical Relationships:

process of understanding the relationships between the variables of a dataset.


```
df = sns.load_dataset("tips")
sns.relplot(x='time',y='tip',data=df,kind='line')
```

2) Plotting with Categorical data:

Main variable is further divided into discrete groups

```
sns.catplot(x='day',y='total_bill',data = df)
```

```
sns.catplot(x='day',y='total_bill',kind ='violin',data = df)
```

```
sns.catplot(x='day',y='total_bill',kind = 'box',data = df)
```

```
sns.catplot(x='day',y='total_bill',kind = 'boxen',data = df)
```

3) Visualizing the distribution of a dataset:

```
from scipy import stats
c=np.random.normal(loc=5,scale=2,size=100)
sns.distplot(c)
```

Multiplotgrids:
 plotted side by side using the same scale and axes to aid the comparison.
 

```
a=sns.load_dataset("iris")
b=sns.FacetGrid(a,col="species")
b.map(plt.hist,"sepal_length")
```

```
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

Plot Aesthetics:

making our plots more attractive and visually delightful

```
sns.set(style="darkgrid")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```
sns.set(style="dark")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```
sns.set(style="white")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```
sns.set(style="white",color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x='day',y='total_bill',data=a)
```

```
sns.set(style="white",color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x='day',y='total_bill',data=a)
sns.despine(offset=10,trim=True)
```

```
c=sns.color_palette()
sns.palplot(c)
```
